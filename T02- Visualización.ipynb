### 1.- Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

$$ f(x) = x^2-4x+2 $$

```python

x = np.linspace(-1, 5, 100)
y1 = x**2 - 4*x + 2

plt.figure(figsize=(6, 4)) 
plt.plot(x, y1, label=r'$f(x) = x^2 - 4x + 2$')
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5)  # Eje y
plt.grid(True)  

plt.title('Gráfico de la función y análisis de su dominio')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

### 2.- Graficar los siguientes grupos de funciones

#### Grupo 1

$$ y6 = 2x+1$$
$$y7=-x+4$$
$$y8=x^2$$

```python
x = np.linspace(-1, 5, 100)
y6 = 2*x + 1
y7 = -x  + 4
y8 = x**2
plt.figure(figsize=(6, 4))
plt.plot(x, y6, label=r'$f(x) = 2x + 1$')
plt.plot(x, y7, label = r'$f(x) = -x + 4$')
plt.plot(x,y8, label = r'$f(x) = x^2$')
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5) # Eje y
plt.grid(True)  

plt.title('Grupo 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

#### Grupo 2

$$ y9 = sin(x) $$
$$ y10 = cos(x) $$

```python
x = np.linspace(-1, 5, 100)
y9 = np.sin(x)
y10 = np.cos(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y9, label=r'$f(x) = sin(x)$')
plt.plot(x, y10, label = r'$f(x) = cos(x)$')
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5) # Eje y
plt.grid(True)  

plt.title('Grupo 2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

### Crea una función que cree una gráfica y dibuje 'k' líneas de forma  $$ f(x) = sin(\alpha x) $$  La función recibe:

```python
def plot_sin_cos_lines(x1, x2, n, a1, a2):
 
    x = np.linspace(x1, x2, n)
    
    alphas = np.linspace(a1, a2, 5)  
    
    plt.figure(figsize=(10, 6))
    
    for alpha in alphas:
        y_sin = np.sin(alpha * x)
        
        plt.plot(x, y_sin, label=r'$f(x) = \sin({:.2f} \cdot x)$'.format(alpha))
        
    plt.axhline(0, color='black', linewidth=0.5)  # Eje x
    plt.axvline(0, color='black', linewidth=0.5)  # Eje y
    plt.grid(True)
    
    plt.title('función seno 2')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    
    plt.show()

plot_sin_cos_lines(-1, 5, 100, 1, 2)
```
